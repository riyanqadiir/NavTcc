Correlation Coefficient

The correlation coefficient is a statistical measure that describes the strength and direction of a linear relationship between two variables. It provides insight into how changes in one variable are associated with changes in another.

Types of Correlation Coefficients:

Pearson Correlation Coefficient (r): Measures the linear relationship between two variables. It ranges from -1 to 1:

r = 1: Perfect positive correlation (both variables increase together).

r = -1: Perfect negative correlation (one variable increases as the other decreases).

r = 0: No linear relationship.

When to Use:

Data is continuous and normally distributed

The relationship between variables is linear

No significant outliers

 Formula:
    r=cov(X,Y)/(σ_X*σ_Y)
​ 
Where:

cov(X, Y) is the covariance

σ_X, σ_Y are the standard deviations of X and Y

Spearman Rank Correlation Coefficient: Measures the strength and direction of the monotonic relationship between two variables, used when the data is not normally distributed or not linear.

Kendall Tau: Another measure of correlation that is based on the ranks of the data. It is less sensitive to tied ranks than Spearman's coefficient.

Formula for Pearson Correlation Coefficient:

r=(n∑xy−∑x∑y)/(n∑x^2−(∑x)^2)(n∑y^2−(∑y)^2)
​
 

​
 
Where:

x and y are the variables.

n is the number of data points.

∑ denotes summation.

```python
# Example of Calculation (Using Python):
# Let's calculate the Pearson Correlation Coefficient between two variables, 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Calculate Pearson Correlation Coefficient
correlation_coefficient, _ = pearsonr(x, y)

# Visualize the data and correlation line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, np.poly1d(np.polyfit(x, y, 1))(x), color='red', label=f'Correlation line (r = {correlation_coefficient:.2f})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Correlation Between X and Y')
plt.legend()
plt.show()

# Print correlation coefficient
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.2f}")
```

Interpretation of the Correlation Coefficient:

1.00: Perfect positive correlation.

-1.00: Perfect negative correlation.

0: No correlation.

0.7 to 1.0: Strong positive correlation.

-0.7 to -1.0: Strong negative correlation.

0.4 to 0.7: Moderate positive correlation.

-0.4 to -0.7: Moderate negative correlation.

Visualizing the Correlation:

In the above example, we use a scatter plot to visualize the relationship between 

x and y, and a red line is drawn to show the linear trend.

Other Types of Correlation (Spearman and Kendall):
For non-linear or non-parametric data, you can use Spearman's rank correlation or Kendall's Tau:



```python
from scipy.stats import spearmanr, kendalltau

# Spearman Rank Correlation
spearman_corr, _ = spearmanr(x, y)
print(f"Spearman Rank Correlation: {spearman_corr:.2f}")

# Kendall Tau Correlation
kendall_corr, _ = kendalltau(x, y)
print(f"Kendall Tau Correlation: {kendall_corr:.2f}")
```

Conclusion:

Pearson is used when the relationship is linear and the data is continuous.

Spearman and Kendall are used for non-parametric data and when the relationship may not be linear but still monotonic.

Univariate, Bivariate, and Multivariate Plots

These types of plots are used to visualize the relationships between one or more variables in a dataset. Let's dive into each one:

A univariate plot is used to visualize the distribution and characteristics of a single variable. It helps to understand the central tendency, spread, and the shape of the data for a single variable.

Common Univariate Plots:

Histogram: Shows the frequency distribution of a variable.

Box Plot: Visualizes the spread and outliers of a single variable.

Density Plot: A smooth version of the histogram that shows the distribution of data.

Example of Univariate Plot:

We will use a Histogram and Box Plot to visualize a single variable:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data for univariate example
data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(data, color='green')
plt.title('Box Plot')

plt.tight_layout()
plt.show()
```

Explanation:

Bivariate Plots

A bivariate plot is used to explore the relationship between two variables. It helps to understand how changes in one variable are associated with changes in another variable.

Common Bivariate Plots:

Scatter Plot: Shows the relationship between two continuous variables.

Hexbin Plot: A variation of the scatter plot used for large datasets.

Pair Plot: A grid of scatter plots between pairs of variables to visualize correlations.

Example of Bivariate Plot:
We will use a Scatter Plot and Pair Plot to visualize the relationship between two variables:

```python
import seaborn as sns
import pandas as pd

# Generate random data for bivariate example
x = np.random.normal(loc=0, scale=1, size=100)
y = 2 * x + np.random.normal(loc=0, scale=0.5, size=100)  # Linear relationship with noise

# Scatter Plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, alpha=0.7, color='red')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Pair Plot (for multivariate case as well)
data = {'X': x, 'Y': y}
df = pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

Explanation:

The Pair Plot shows scatter plots of multiple pairs of variables (it can also be extended to multivariate data).

Multivariate Plots

A multivariate plot is used when you want to analyze more than two variables and understand how multiple variables relate to each other. These plots are helpful for identifying interactions and correlations among several variables.

Common Multivariate Plots:

Pair Plot (also multivariate): Displays scatter plots for pairs of variables and can include histograms or density plots on the diagonal for each variable.

Heatmap: Shows the correlation between multiple variables.

3D Scatter Plot: Adds a third dimension (Z) to the scatter plot for visualizing three variables.

Facet Grid: Creates a grid of plots based on the values of one or more categorical variables.

Example of Multivariate Plot:

We will use a Pair Plot, Heatmap, and 3D Scatter Plot:

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate random data for multivariate example
data = np.random.normal(size=(100, 3), loc=0, scale=1)  # Three variables
df = pd.DataFrame(data, columns=['X', 'Y', 'Z'])

# Pair Plot
sns.pairplot(df)
plt.suptitle('Pair Plot for Multivariate Data', y=1.02)
plt.show()

# Heatmap (Correlation Matrix)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# 3D Scatter Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], color='orange')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Scatter Plot')
plt.show()
```

Explanation:

The Pair Plot creates scatter plots for every combination of the three variables and can help to detect relationships between them.

The Heatmap visualizes the correlation matrix, showing how strongly variables are related (the closer to 1 or -1, the stronger the correlation).

The 3D Scatter Plot visualizes the relationship between three variables in a 3D space.

Probability:

Probability is a branch of mathematics that deals with the likelihood of events occurring. It is a measure of the chance that a particular outcome will happen, expressed as a number between 0 and 1.

0 indicates that an event will not occur.

1 indicates that an event will certainly occur.

Probabilities of all possible outcomes of an experiment sum to 1.

Basic Terminology in Probability:

Experiment: A procedure that yields one outcome from a set of possible outcomes (e.g., rolling a die).

Sample Space (S): The set of all possible outcomes of an experiment (e.g., for rolling a fair six-sided die, the sample space is 
}
S={1,2,3,4,5,6}).

Event (E): A subset of the sample space. It represents a specific outcome or a set of outcomes (e.g., rolling an even number: 

E={2,4,6}).

Outcome: The result of a single trial of an experiment (e.g., getting a 3 when rolling a die).

Complementary Event: The event that something will not happen (e.g., the complement of rolling an even number is rolling an odd number).

Example 1: Calculating Probability of Rolling a Die

Let's calculate the probability of rolling a number greater than 4 on a six-sided die:

Sample space 

S={1,2,3,4,5,6}

Favorable outcomes 

E={5,6}

Probability 

P(E)= 2/6=0.33

Conditonal Probability:

Let’s say you have a deck of 52 playing cards.

Event A: Getting a King

Event B: Getting a red card (there are 26 red cards: hearts and diamonds)

Now, if I tell you that a red card has already been drawn, what is the probability that it’s a King?

Only 2 red Kings exist (King of hearts and King of diamonds)
Total red cards = 26

So,

P(King∣Red)= 2/26= 1/13

This is a conditional probability—you’re not considering the whole deck anymore, just the 26 red cards.

Example With Dice
Let’s say you roll a 6-sided die.

Event A: Rolling a number greater than 3 → {4, 5, 6}

Event B: Rolling an even number → {2, 4, 6}

What’s the probability of rolling a number greater than 3 given that the number is even?

A∩B = Numbers that are both >3 and even = {4, 6}
Total even numbers = {2, 4, 6} → 3 numbers

So,

P(A∣B)= 2/3

```python
import numpy as np

# Sample space and favorable outcomes for a six-sided die
sample_space = [1, 2, 3, 4, 5, 6]
even_numbers = [2, 4, 6]

# Probability of rolling an even number
P_even = len(even_numbers) / len(sample_space)
print(f"Probability of rolling an even number: {P_even}")

# Probability of rolling a 3 or 5
P_3_or_5 = (1/6) + (1/6)  # Since these events are mutually exclusive
print(f"Probability of rolling either a 3 or 5: {P_3_or_5}")

# Conditional Probability example
# Total number of red cards (Hearts + Diamonds)
total_red_cards = 26

# Number of Kings that are red (King of Hearts and King of Diamonds)
red_kings = 2

# Conditional Probability: P(King | Red)
conditional_probability = red_kings / total_red_cards

# Display result
print("P(King | Red) =", conditional_probability)
```

Explanation of Code:

P_even: This is the probability of rolling an even number (either 2, 4, or 6) on a six-sided die.

P_3_or_5: This is the probability of rolling either a 3 or a 5 on the die. Since these are mutually exclusive events, the probability is the sum of the individual probabilities.

Joint, Marginal, and Conditional Probability

In probability theory, joint, marginal, and conditional probability are concepts that help us understand the relationships between two or more events in a probability space. These concepts are essential in understanding how events interact with each other, especially in fields like statistics and machine learning.

Joint Probability:

The joint probability of two events A and B is the probability that both events occur simultaneously. It is denoted as P(A∩B), and it answers the question: "What is the probability that both event A and event B happen?"

The formula for joint probability is:

P(A∩B)=P(A and B)

Example of Joint Probability:

Suppose we have a deck of cards, and we want to find the probability of drawing a red card (event A) and a face card (event B) simultaneously.

Total number of cards in a deck = 52

Red cards (hearts and diamonds) = 26

Face cards (Jack, Queen, King) = 12

Red face cards (hearts and diamonds face cards) = 6 (Jack, Queen, King of hearts and diamonds)

The joint probability of drawing a red face card is:

P(Red Face)= 6/52=0.115

Python Code Example for Joint Probability:

```python
# Total number of cards in a deck
total_cards = 52

# Number of red face cards (Jack, Queen, King of hearts and diamonds)
red_face_cards = 6

# Joint probability of drawing a red face card
P_red_face = red_face_cards / total_cards
print(f"Joint Probability of drawing a red face card: {P_red_face}")
```

Marginal Probability:

The marginal probability is the probability of an event occurring irrespective of the outcome of another event. It is calculated by summing or integrating over all possible outcomes of the other variable(s).

For two events A and B, the marginal probability of A, denoted as P(A), is calculated by summing the joint probability over all possible outcomes of event B (or vice versa).

The formula for marginal probability is:

P(A)= ∑P(A∩B)

Or for a continuous case:

P(A)=∫P(A∩B)dB
Example of Marginal Probability:

If we want the probability of drawing a red card (event A) from the deck of cards, we calculate the marginal probability by summing over all possible outcomes for face cards and non-face cards:

Red cards = 26

Total cards = 52

The marginal probability of drawing a red card is:

P(Red)= 26/52=0.5

Python Code Example for Marginal Probability:

```python
# Number of red cards in a deck
red_cards = 26

# Marginal probability of drawing a red card
P_red = red_cards / total_cards
print(f"Marginal Probability of drawing a red card: {P_red}")
```

Conditional Probability

Conditional probability is the probability of event A occurring given that event B has already occurred. It is denoted by P(A∣B), and it answers the question: "What is the probability of event A happening given that event B has happened?"

The formula for conditional probability is:

P(A∣B)= P(A∩B)/P(B)
​
 
Where:

P(A∩B) is the joint probability of events 

P(B) is the marginal probability of event 𝐵.

Example of Conditional Probability:

Let’s assume we draw one card from the deck of cards, and we are told that the card is red. What is the probability that the card is a face card, given that it is red?

Red cards (event A) = 26

Red face cards (event A∩B) = 6

Probability of drawing a red card (marginal probability of A) = 26/52=0.5

Conditional probability of drawing a red face card given that the card is red is:

P(Face|Red)= P(Red Face)/P(Red)= (6/52)/(26/52)=6/26=0.23
Python Code Example for Conditional Probability:

```python
# Probability of drawing a red card
P_red = red_cards / total_cards

# Conditional probability of drawing a face card given that the card is red
P_face_given_red = red_face_cards / red_cards
print(f"Conditional Probability of drawing a face card given that the card is red: {P_face_given_red}")
```

Probability Distributions

A probability distribution describes how the values of a random variable are distributed. It provides a function that assigns probabilities to different outcomes. There are two main types:

Discrete Probability Distributions – Used for discrete random variables (finite/countable outcomes).

Continuous Probability Distributions – Used for continuous random variables (infinite possible values).

Discrete Probability Distributions

A discrete probability distribution gives the probability of each possible value of a discrete random variable.

A. Bernoulli Distribution

Represents a single trial with two possible outcomes: Success (1) or Failure (0).

Example: Tossing a coin where Head = 1, Tail = 0.

Probability Mass Function (PMF):

P(X=x)=p^x(1−p)^(1−x) ,x∈{0,1}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.5  # Probability of success (Head)
x = [0, 1]
y = bernoulli.pmf(x, p)

plt.bar(x, y, color=['blue', 'red'])
plt.xticks(x, ['Tail (0)', 'Head (1)'])
plt.ylabel('Probability')
plt.title('Bernoulli Distribution (p=0.5)')
plt.show()
```

Binomial Distribution

Models the number of successes in n independent Bernoulli trials.

Example: Number of heads in 10 coin flips.

PMF:

P(X=k)=(n
        k)p^k(1−p)^(n−k)  ,k=0,1,2,…,n

```python
from scipy.stats import binom

n, p = 10, 0.5  # 10 coin flips, probability of heads = 0.5
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

plt.bar(x, y, color='purple')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

Poisson Distribution

Models the number of events occurring in a fixed interval (time, area, volume) given a constant average rate λ.

Example: Number of calls received in a call center per hour.

PMF:

P(X=k)= λ^k e^−λ/k! ,k=0,1,2,3,…

```python
from scipy.stats import poisson

λ = 4  # Average rate of occurrences
x = np.arange(0, 15)
y = poisson.pmf(x, λ)

plt.bar(x, y, color='orange')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ=4)')
plt.show()
```

Continuous Probability Distributions

A continuous probability distribution assigns probabilities to intervals rather than discrete values.

A. Uniform Distribution

Every value in an interval has an equal probability.

Example: Rolling a fair die (1 to 6).

Probability Density Function (PDF):

f(x)= 1/(b−a) ,a≤x≤b

```python
from scipy.stats import uniform

a, b = 0, 10  # Range from 0 to 10
x = np.linspace(a, b, 100)
y = uniform.pdf(x, a, b-a)

plt.plot(x, y, color='green')
plt.fill_between(x, y, alpha=0.4)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Uniform Distribution (0 to 10)')
plt.show()
```

 Normal (Gaussian) Distribution

Bell-shaped curve where most values cluster around the mean.

Example: Heights of people, test scores.

PDF:

f(x)= (1/(σ sqrt(2π)))e^−(x−μ)^2/2σ^2

```python
from scipy.stats import norm

mu, sigma = 0, 1  # Mean = 0, Standard Deviation = 1
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y, color='blue')
plt.fill_between(x, y, alpha=0.4)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution (Mean=0, SD=1)')
plt.show()
```

Exponential Distribution

Models the time between events in a Poisson process.

Example: Time until next earthquake, time between customer arrivals.

PDF:

f(x)=λe^−λx ,x≥0

```python
from scipy.stats import expon

λ = 1
x = np.linspace(0, 5, 100)
y = expon.pdf(x, scale=1/λ)

plt.plot(x, y, color='red')
plt.fill_between(x, y, alpha=0.4)
plt.xlabel('Time')
plt.ylabel('Density')
plt.title('Exponential Distribution (λ=1)')
plt.show()
```

Bayesian Probability

Bayesian probability is an approach to probability based on Bayes' Theorem, which updates our beliefs as we gain new information. It is widely used in machine learning, data science, and decision-making.

Bayes’ Theorem

Bayes' theorem provides a way to update probabilities based on new evidence:

P(A∣B)= P(B∣A)P(A)/P(B)
​
Where:

P(A∣B) = Posterior Probability (Updated probability of A given 𝐵

P(B∣A) = Likelihood (Probability of observing 𝐵 given 𝐴

P(A) = Prior Probability (Initial probability of A before observing 𝐵

P(B) = Marginal Probability (Total probability of B happening)

Example: Medical Testing for a Disease
Let's say:

1% of the population has a disease (P(D)=0.01).

A test is 90% accurate for sick people (P(T∣D)=0.9).

The test gives a false positive rate of 5% (P(T∣¬D)=0.05).

Now, if a person tests positive, what is the probability they actually have the disease?

Using Bayes’ theorem:

P(D∣T)= P(T∣D)P(D)/P(T)
​
Where:

P(T)=P(T∣D)P(D)+P(T∣¬D)P(¬D)

P(T)=(0.9×0.01)+(0.05×0.99)=0.009+0.0495=0.0585

Now:

P(D∣T)= 0.9×0.01/0.0585=0.154

So even with a positive test, the probability of actually having the disease is only 15.4%, not 90%!

```python
# Given probabilities
P_D = 0.01  # Prior (disease probability)
P_T_given_D = 0.9  # Likelihood (test positive given disease)
P_T_given_not_D = 0.05  # False positive rate
P_not_D = 1 - P_D  # Probability of not having the disease

# Total probability of testing positive
P_T = (P_T_given_D * P_D) + (P_T_given_not_D * P_not_D)

# Bayes' Theorem: Posterior probability
P_D_given_T = (P_T_given_D * P_D) / P_T

print(f"Probability of having the disease given a positive test: {P_D_given_T:.4f}")
```

Applications of Bayesian Probability

Bayesian probability is widely used in:

Spam Filtering – Classifying emails as spam or not using prior probabilities.

Machine Learning – Bayesian classifiers like Naive Bayes.

Medical Diagnosis – Updating the likelihood of diseases based on symptoms.

Stock Market Predictions – Adjusting predictions based on new data.

AI & Robotics – Probabilistic reasoning in self-driving cars and robotics.
