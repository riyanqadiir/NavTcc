What is Machine Learning?
Machine Learning is a branch of Artificial Intelligence (AI) that enables computers to learn from data and make decisions or predictions without being explicitly programmed.

Instead of writing rules manually, we train a model using data ‚Äî the model learns patterns and relationships and uses them to make future predictions.

Types of Machine Learning

There are three main types of machine learning:

1. Supervised Learning
Definition:
The model learns from a labeled dataset, which means the input comes with the correct output.

Goal:
Predict the output for new, unseen data based on past labeled data.

Examples:

Predicting house prices (input: size, location; output: price)

Diagnosing diseases (input: symptoms; output: disease)

Common algorithms:

Linear Regression

Logistic Regression

Decision Trees

Support Vector Machines (SVM)

Random Forest

k-Nearest Neighbors (KNN)

2. Unsupervised Learning

Definition:

The model learns from unlabeled data ‚Äî it tries to find hidden patterns or groupings.

Goal:
Discover the structure of the data ‚Äî such as clusters or anomalies.

Examples:

Customer segmentation (grouping similar customers)

Anomaly detection in credit card transactions

Common algorithms:

K-Means Clustering

Hierarchical Clustering

DBSCAN

PCA (Principal Component Analysis)

3. Reinforcement Learning

Definition:

An agent learns by interacting with an environment. It takes actions, receives rewards or penalties, and learns a policy to maximize the total reward.

Goal:

Learn the best actions to take in a given environment.

Examples:

Game playing (e.g., AlphaGo)

Self-driving cars (learning how to drive safely)

Robotics (balancing or walking)

Common algorithms:

Q-learning

Deep Q-Networks (DQN)

Policy Gradient Methods

1. Supervised Learning ‚Äì Predicting House Prices

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd

# Sample data: Size (sqft) and Price ($1000s)
data = pd.DataFrame({
    'size': [500, 750, 1000, 1250, 1500],
    'price': [150, 200, 250, 275, 300]
})

X = data[['size']]  # Input feature
y = data['price']   # Target

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
prediction = model.predict([[1100]])
print(f"üè† Predicted price for 1100 sqft: ${prediction[0]*1000:.2f}")
```

2. Unsupervised Learning ‚Äì Clustering Customers with KMeans

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Sample data: Annual Income and Spending Score
X = np.array([
    [15, 39],
    [16, 81],
    [17, 6],
    [18, 77],
    [19, 40],
    [20, 76]
])

# KMeans Clustering (k=2)
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_

# Plotting clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("üõçÔ∏è Customer Segments")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

3. Reinforcement Learning (Simple Simulation) ‚Äì Q-learning-like Grid Example
Reinforcement learning usually requires custom environments. Here's a simple simulation using rewards and actions:

```python
import numpy as np

# Simple grid: 0 = empty, 1 = goal
grid = np.array([
    [0, 0, 1],
    [0, -1, 0],
    [0, 0, 0]
])

# Agent at bottom-left
position = [2, 0]

# Rewards
rewards = {
    (0, 2): 10,  # Goal
    (1, 1): -5   # Danger
}

def step(pos, action):
    """Move agent in the grid"""
    x, y = pos
    if action == "up": x -= 1
    elif action == "down": x += 1
    elif action == "left": y -= 1
    elif action == "right": y += 1

    # Stay in bounds
    x = max(0, min(2, x))
    y = max(0, min(2, y))

    new_pos = [x, y]
    reward = rewards.get(tuple(new_pos), 0)
    return new_pos, reward

# Try moving right twice, then up
pos, r1 = step(position, "right")
pos, r2 = step(pos, "right")
pos, r3 = step(pos, "up")
print(f"ü§ñ Final Position: {pos}, Total Reward: {r1 + r2 + r3}")
```

```python
import pandas as pd

# Example: Load dataset
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
print(data.head())
data['species'].unique()
data['species'].values
```

2. Data Preprocessing
Involves cleaning the data:

Handling missing values

Encoding categorical data

Normalizing or scaling

Data Preprocessing in Machine Learning

Data preprocessing is a critical step that transforms raw data into a format that machine learning algorithms can work with effectively. Proper preprocessing can significantly improve model performance.

Key Steps in Data Preprocessing

1. Data Cleaning

Handling missing values:

Deletion (remove rows/columns with missing values)

Imputation (mean, median, mode, or predictive imputation)

Flagging (mark missing values as a separate category)

Outlier detection and treatment:

Z-score or IQR methods for detection

Capping, transformation, or removal

2. Data Transformation

Normalization (scaling to [0,1] range):

Log/Power transformations (for skewed data):

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical target variable
le = LabelEncoder()
data['species'] = le.fit_transform(data['species'])
print(data['species'].values)
# Check for missing values
print(data.isnull().sum())
```

3. Feature Crafting (Feature Engineering)

This includes selecting or creating features that improve model performance.

```python
# Example: Create a new feature (petal ratio)
data['petal_ratio'] = data['petal_length'] / data['petal_width']
```

4. Modeling (Training)
Choosing and training a machine learning model using training data.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = data.drop('species', axis=1)
y = data['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

5. Testing and Evaluation

Evaluate the model's performance using metrics like accuracy, precision, recall, F1-score, or confusion matrix.

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

6. Deployment
This step involves making the model accessible via a web app, API, or embedding it into a mobile/desktop app.

Using joblib to save and reuse model:

```python
import joblib

# Save model
joblib.dump(model, 'iris_model.pkl')
```

```python
# Later load it
model = joblib.load('iris_model.pkl')
```
