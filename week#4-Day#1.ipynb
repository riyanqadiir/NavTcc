Merging / Joining in pandas

```python
import pandas as pd
df_1=pd.DataFrame({
    'id':[1,2,3,4],
    'names':['Alice', 'Bob', 'Charlie', 'David']
}
)
df_2=pd.DataFrame({
    'id':[3,4,5,6],
    'score':[85,90,95,80]
}
)
```

INNER JOIN: Only matching rows in both

```python
result_inner = pd.merge(df_1, df_2, on='id', how='inner')
print(result_inner)
```

LEFT JOIN: All from df_1, matching from df_2

```python
result_left = pd.merge(df_1, df_2, on='id', how='left')
print(result_left)
```

RIGHT JOIN: All from df_2, matching from df_1

```python
result_right = pd.merge(df_1, df_2, on='id', how='right')
print(result_right)
```

OUTER JOIN: All rows from both, NaN where no match

```python
result_outer = pd.merge(df_1, df_2, on='id', how='outer')
print(result_outer)
```

Concatenation in pandas
Used to stack DataFrames vertically or horizontally.

```python
df_a = pd.DataFrame({'id': [1, 2], 'name': ['Alice', 'Bob']})
df_b = pd.DataFrame({'id': [3, 4], 'name': ['Charlie', 'David']})
```

Concatenate vertically (default):

```python
result_concat = pd.concat([df_a, df_b])
print(result_concat)
```

Concatenate horizontally (axis=1):

```python
result_concat_col = pd.concat([df_a, df_b], axis=1)
print(result_concat_col)
```

GroupBy in pandas

```python
import pandas as pd

# Sample Data
data = {
    'department': ['IT', 'HR', 'IT', 'HR', 'Finance', 'Finance'],
    'employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'salary': [70000, 50000, 80000, 45000, 75000, 78000]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Group by department and calculate mean salary
grouped = df.groupby('department')['salary'].mean()
print("ðŸ“Š Grouped by department:\n", grouped)
```

```python
import matplotlib.pyplot as plt

# For plots inside Jupyter
%matplotlib inline

# Sample numeric DataFrame
df_plot = pd.DataFrame({
    'A': [1, 3, 4, 6, 8],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 3, 6, 2, 1]
})
```

An area plot displays quantitative data graphically. It is based on the line plot, but fills the area beneath the lines to emphasize the magnitude of change over time.

```python
# Area plot
df_plot.plot.area(title="Area Plot")
plt.show()
```

A bar plot is used to visualize categorical data by representing each category as a bar with a height or length corresponding to its value.

```python
# Bar plot
df_plot.plot.bar(title="Bar Plot")
plt.show()
```

```python
# Horizontal bar
df_plot.plot.barh(title="Horizontal Bar Plot")
plt.show()
```

A line plot is used to represent numerical data that changes continuously over time or sequence, showing how one variable is affected by another.

A line plot (or line chart) is a type of chart used to display data points over a continuous interval or time period. It connects data points with a straight line, helping to visualize trends, changes, or patterns in data.

```python
# Line plot (default)
df_plot.plot(title="Line Plot")
plt.show()
```

A histogram is a graphical representation used to visualize the distribution of a dataset. It groups continuous numerical data into intervals (called bins) and then counts how many data points fall into each bin.

```python
# Histogram
df_plot.plot.hist(title="Histogram", alpha=0.7)
plt.show()
```

A box plot (also called a box-and-whisker plot) is a graphical summary of a dataset that shows its distribution, central tendency, and variability. It is especially useful for detecting outliers and comparing distributions across groups.

Key Components of a Box Plot:

Box:

Represents the interquartile range (IQR) â€” the middle 50% of the data.

Bottom edge of the box = Q1 (25th percentile)

Top edge of the box = Q3 (75th percentile)

Line inside the box = Median (Q2 or 50th percentile)

Whiskers:

Lines extending from the box to the minimum and maximum values within 1.5 * IQR from Q1 and Q3.

Help identify the spread of the rest of the data.

Outliers:

Data points outside 1.5 * IQR from Q1 or Q3.

Often plotted as individual dots or stars.

Use Cases:

Visualizing the distribution of numerical data

Comparing distributions across different categories or groups

Spotting skewness or outliers quickly



```python
# Box plot
df_plot.plot.box(title="Box Plot")
plt.show()
```

A density plot (also called a Kernel Density Estimate (KDE) plot) is a smooth, continuous estimate of the distribution of a dataset. It is a useful way to visualize the probability distribution of a continuous random variable, providing an alternative to a histogram. A density plot gives more insight into the shape of the distribution, especially when comparing multiple distributions.

```python
# Density (KDE) plot
df_plot.plot.kde(title="KDE Plot")
plt.show()
```

A scatter plot is a type of data visualization used to display the relationship between two continuous variables. It uses dots to represent individual data points, where each dot's position on the horizontal axis (x-axis) corresponds to one variable, and its position on the vertical axis (y-axis) corresponds to another variable. Scatter plots are great for visualizing correlations, trends, and patterns in data.

```python
# Scatter plot (columns A vs B)
df_plot.plot.scatter(x='A', y='B', title="Scatter Plot")
plt.show()
```

A hexbin plot is a type of data visualization used to represent the relationship between two continuous variables. It is similar to a scatter plot, but instead of displaying individual data points, it divides the plot into a grid of hexagonal bins (or cells), where each hexagon represents a range of values. The color of each hexagon typically indicates the number of data points that fall within that bin. Hexbin plots are especially useful when visualizing large datasets, as they help reduce the clutter caused by overlapping points in a scatter plot.

Key Features of a Hexbin Plot:

Hexagonal Bins: The plot divides the area into hexagonal shapes, unlike the rectangular bins used in 2D histograms.

Color Mapping: The color of each hexagon represents the density of data points within that bin. Typically, darker colors indicate higher density, while lighter colors represent lower density.

Two Continuous Variables: The plot is used to explore the relationship between two continuous variables.

Use Cases:

Visualizing Large Datasets: Hexbin plots are ideal when you have large datasets with many points that might overlap in a scatter plot, leading to visual clutter.

Understanding Density: Hexbin plots help visualize the density of data points in various regions of the plot, which is useful for identifying areas of high concentration.



```python
# Hexbin (for dense data)
df_plot.plot.hexbin(x='A', y='B', gridsize=25, title="Hexbin Plot")
plt.show()
```
